## Creating defog dataframe using pandas

```python
# Reading all the csv files in the dataframe
#file_path = "/kaggle/input/tlvmc-parkinsons-freezing-gait-prediction/train/defog"
file_path = "/kaggle/input/tlvmc-parkinsons-freezing-gait-prediction/train/defog/139f60d29b.csv"
df_defog = pd.read_csv(file_path,index_col = None, header = 0)

#csv_files = glob.glob(os.path.join(file_path,"*.csv"))

#file_list = []
#for file_name in csv_files:
 #   df = pd.read_csv(file_name,index_col = None, header = 0)
 #   file_list.append(df)
#df_defog = pd.concat(file_list, axis=0, ignore_index=True)  
```

##  Checking if there are any missing values in defog dataset 

```python
# Checking if any of the dataset columns have missing values
df_defog.isnull().values.any().sum()
```

## Describing the defog datasets with skiwness and kurtosis

```python
# Describing the datasets with variance, skewness and kurtosis
stats_defog=df_defog.describe(include = 'all')
stats_defog.loc['var'] = df_defog.var().tolist()
stats_defog.loc['skew'] = df_defog.skew().tolist()
stats_defog.loc['kurt'] = df_defog.kurtosis().tolist()
non_nan_df_defog=stats_defog.fillna(0)
```

```python
# Transposing the describe dataframe to convert rows to columns
non_nan_df_defog.transpose().style.background_gradient(subset = ["count","unique","top","freq","mean","std","min","25%","50%","75%","max","var","kurt","skew"], 
                             cmap = "seismic", 
                             vmin = -1, 
                             vmax = 1)
```

## Correlation matrix among Time, AccV,AccML and AccAP 

```python
# Finding the correlation matrix and excluding the varaible fields
corr_defog_df=np.round(non_nan_df_defog.drop(
    ["StartHesitation", "Walking","Turn","Valid","Task"], axis=1
).corr(), 3)
```

```python
# Creating correlation matrix using plotly.
corr_matrix = px.imshow(
    corr_defog_df,
    x=corr_defog_df.columns,
    y=corr_defog_df.columns, 
    text_auto=True
   )
corr_matrix.update_xaxes(side="bottom")
corr_matrix.update_layout(
    title={
        'text': "Correlation Matrix of defog train data",
        'y':0.95,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    }
)

corr_matrix.show()
```
