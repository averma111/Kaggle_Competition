```python
# Creating DEFOG datasets using all the csv files 

root_file_path='/kaggle/input/tlvmc-parkinsons-freezing-gait-prediction/train/defog/'
DEFOG = pd.DataFrame()
for root, dirs, files in os.walk(root_file_path):
    for name in files:       
        f = os.path.join(root, name)
        df_list= pd.read_csv(f)
        words = name.split('.')[0]
        df_list['file']= name.split('.')[0]
        DEFOG = pd.concat([DEFOG, df_list], axis=0)
        

DEFOG
```

```python
# Creating TDCSFOG datasets using all the csv files 

root_file_path='/kaggle/input/tlvmc-parkinsons-freezing-gait-prediction/train/tdcsfog'
TDCSFOG = pd.DataFrame()
for root, dirs, files in os.walk(root_file_path):
    for name in files:       
        f = os.path.join(root, name)
        df_list= pd.read_csv(f)
        words = name.split('.')[0]
        df_list['file']= name.split('.')[0]
        TDCSFOG = pd.concat([TDCSFOG, df_list], axis=0)
        

TDCSFOG
```

```python
# Using DEFOG valid data for further analysis
DEFOG = DEFOG[(DEFOG['Task']==1) & (DEFOG['Valid']==1)]
```

```python
# Reading defog metadata file and combining it with DEFOG dataset
defog_metadata = pd.read_csv("/kaggle/input/tlvmc-parkinsons-freezing-gait-prediction/defog_metadata.csv")
defog_metadata
```

```python
# Merging the DEFOG and defog_metadata datasets into one dataset
defog_merged= defog_metadata.merge(DEFOG, how = 'inner', left_on = 'Id', right_on = 'file')
defog_merged.drop(['file','Valid','Task'], axis = 1, inplace = True)
defog_merged
```

```python
# Reading tdcsfog_metadata dataset 
tdcsfog_metadata = pd.read_csv("/kaggle/input/tlvmc-parkinsons-freezing-gait-prediction/tdcsfog_metadata.csv")
tdcsfog_metadata
```

```python
# Merging the TDCSFOG and tdcsfog_metadata datasets into one dataset
tdcsfog_merged= tdcsfog_metadata.merge(TDCSFOG, how = 'inner', left_on = 'Id', right_on = 'file')
tdcsfog_merged.drop(['file'], axis = 1, inplace = True)
tdcsfog_merged
```

## Feature Engineering of Time Series data

```python
# Importing training models packages
from sklearn.model_selection import KFold, StratifiedKFold, train_test_split, GridSearchCV
import warnings
warnings.filterwarnings('ignore')
```

```python
conditions = [
    (defog_merged['StartHesitation'] == 1),
    (defog_merged['Turn'] == 1),
    (defog_merged['Walking'] == 1)]
choices = ['StartHesitation', 'Turn', 'Walking']
defog_merged['event'] = np.select(conditions, choices, default='Normal')
```

```python
# Creating the traning dataframe
train_df = defog_merged[['AccV','AccML','AccAP','event']]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

train_df['target'] = le.fit_transform(train_df['event'])
```

```python
X = train_df.drop(['event','target'], axis=1)
y = train_df['target']
```

## Traning the LIGHTBGM model

```python
import lightgbm as lgb


# split dataset into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1004)

#Converting the dataset in proper LGB format
d_train=lgb.Dataset(X_train, label=y_train)
#setting up the parameters
params={}
params['learning_rate']=0.03
params['boosting_type']='gbdt' #GradientBoostingDecisionTree
params['objective']='multiclass' #Multi-class target feature
params['metric']='multi_logloss' #metric for multi-class
params['max_depth']=7
params['num_class']=4 #no.of unique values in the target class not inclusive of the end value
params['verbose']=-1
#training the model
clf=lgb.train(params,d_train,1000)  #training the model on 1,000 epocs
#prediction on the test dataset
y_pred_1=clf.predict(X_test)
```

```python
from sklearn.metrics import precision_score
precision_score(y_test, np.argmax(y_pred_1, axis=-1), average='macro')
```
