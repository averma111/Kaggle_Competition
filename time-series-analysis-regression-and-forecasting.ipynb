```python
# Importing necessary variables
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Reading the dataframe
df = pd.read_csv("/kaggle/input/autocityhighway/uciml_auto_city_highway_mpg.csv")
```

```python
df.shape
```

```python
df.head(5)
```

```python
# Plotting the original dataset
df.plot.scatter(x="City MPG",y="Highway MPG")
plt.show()
```

```python
X = df.iloc[:,0:1].values
y = df.iloc[:,1:2].values
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state = 40)
```

```python
LR = LinearRegression()
LR.fit(X_train,y_train)
```

```python
# print out the coorelation coefficient for the training dataset
print('r='+str(LR.score(X_train, y_train)))
```

```python
# Plot the regression line superimposed on the training dataset
plt.xlabel('City MPG')
plt.ylabel('Highway MPG')
plt.scatter(X_train, y_train, color = 'blue')
plt.plot(X_train, LR.predict(X_train), color = 'black')
plt.show()
```

```python
# Plot the predicted and actual values for the holdout dataset
plt.xlabel('City MPG')
plt.ylabel('Highway MPG')
actuals = plt.scatter(X_test, y_test, marker='o', color = 'lightblue', label='Actual values')
predicted = plt.scatter(X_test, LR.predict(X_test), marker='+', color = 'black', label='Predicted values')
plt.legend(handles=[predicted, actuals])
plt.show()
```
